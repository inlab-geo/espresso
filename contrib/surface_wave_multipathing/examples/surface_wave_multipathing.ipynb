# Espresso Interface to surface wave multipathing

This Jupyter notebook allows to test the surface wave mutlipathing examples exposed through the Espresso Interface. Espresso is used as a simplified interface to the swmp package avaialble here https://github.com/JuergHauser/swmp

Later arrivals are a function of the velocity model and as consequence thereof the number of predicitons is unlikely to be equal to the number of observations. The simplest solution to adress this is to limit the predictions and observations to the subset of observations and predictions that are predicted and observed. This subset can be accessed via the `joint_data_and_jacobian` function of this problem.
